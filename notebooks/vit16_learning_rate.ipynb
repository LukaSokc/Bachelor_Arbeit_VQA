# Erstes Modell vit16 und biobert, mit lr 1e4 und 5 epochen und batch size 8

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV-Datei laden (falls als Datei vorhanden, sonst direkt als DataFrame einlesen)
csv_file = "../docs/train_batch_logs.csv"  # Ersetze dies mit deinem Dateinamen
df = pd.read_csv(csv_file)

# Anzahl der Epochen herausfinden
epochen = df["epoch"].unique()

# Plots erstellen
plt.figure(figsize=(12, 8))

for i, epoch in enumerate(epochen[:5], 1):  # Maximal 5 Epochen plotten
    epoch_data = df[df["epoch"] == epoch]
    epoch_data = epoch_data[epoch_data["batch_idx"] % 10 == 0]  # Nur jeden 10. Schritt nehmen
    
    plt.subplot(3, 2, i)  # 3 Zeilen, 2 Spalten Layout
    plt.plot(epoch_data["batch_idx"], epoch_data["train_loss"], linestyle="-", label=f"Epoche {epoch}")
    plt.xlabel("Batch Index")
    plt.ylabel("Train Loss")
    plt.title(f"Train Loss - Epoche {epoch}")
    plt.legend()
    plt.grid()

plt.tight_layout()
plt.show()
```
