```python
from datasets import load_from_disk
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
project_root = Path.cwd().parent
data_path_test = project_root / "data" / "test"
data_path_validation = project_root / "data" / "validation"
data_path_train = project_root / "data" / "train"

dataset = load_from_disk(str(data_path_test))
```

```python
print(dataset[0])
```

```python
sample = dataset[10]  
image = sample["image"] 
question = sample["question"]
answer = sample["answer"]

plt.imshow(image)
plt.axis("off") 
plt.title(f"Q: {question}\nA: {answer}", fontsize=10)
plt.show()
```

# Längste Frage und Antwort aus dem Testset um zu bestimmen wie gross die Tokensize sein soll

```python
max_question = max(dataset, key=lambda x: len(x["question"].split()))
max_answer = max(dataset, key=lambda x: len(x["answer"].split()))

print("Längste Frage:", max_question["question"])
print("Wörter:", len(max_question["question"].split()))
print("Zeichen:", len(max_question["question"]))

print("\nLängste Antwort:", max_answer["answer"])
print("Wörter:", len(max_answer["answer"].split()))
print("Zeichen:", len(max_answer["answer"]))
```
