```python
from pathlib import Path
from datasets import load_from_disk
import matplotlib.pyplot as plt

# Pfad zum Dataset definieren
project_root = Path.cwd().parent
data_path = project_root / "data" / "train"

# Dataset laden
dataset = load_from_disk(str(data_path))

# Beispielhafte Struktur der Daten anzeigen
print("Dataset Keys:", dataset.features.keys())
print("Anzahl der Eintrge:", len(dataset))

# Ein einzelnes Sample inspizieren
sample = dataset[0]  # Erstes Sample nehmen

# Struktur des ersten Datensatzes ausgeben
for key, value in sample.items():
    print(f"{key}: {type(value)}")
    if isinstance(value, str) or isinstance(value, int):
        print(f"  Beispielwert: {value}")
    elif isinstance(value, list):
        print(f"  Beispielwert (erste 3 Elemente): {value[:3]}")
    elif isinstance(value, dict):
        print(f"  Keys: {value.keys()}")
    elif hasattr(value, 'shape'):
        print(f"  Shape: {value.shape}")

# Falls Bilder enthalten sind, ein Beispielbild anzeigen
if "image" in sample:
    image = sample["image"]
    plt.imshow(image)
    plt.axis("off")
    plt.title("Beispielbild aus dem Dataset")
    plt.show()
```
