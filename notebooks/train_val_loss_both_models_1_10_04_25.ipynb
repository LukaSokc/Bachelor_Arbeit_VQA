```python
import matplotlib.pyplot as plt

# Schritte (gemeinsam für beide Modelle)
steps = list(range(1000, 10000, 1000))

# Qwen2.5VL 3B Daten
training_loss_qwen = [6.1595, 5.7068, 5.6267, 5.6446, 5.6287, 5.5940, 5.6683, 5.6096, 5.6435]
validation_loss_qwen = [5.817586, 5.682501, 5.656390, 5.641116, 5.635643, 5.633337, 5.630905, 5.629152, 5.627742]

# Gemma 3B Daten
training_loss_gemma = [0.3246, 0.2845, 0.2672, 0.2585, 0.2402, 0.2274, 0.2127, 0.2121, 0.2038]
validation_loss_gemma = [0.311775, 0.292898, 0.283513, 0.276960, 0.279569, 0.277382, 0.267470, 0.269880, 0.268646]

# Plot 1: Training Loss
plt.figure(figsize=(10, 5))
plt.plot(steps, training_loss_qwen, marker='o', label='Qwen2.5VL 3B - Training Loss')
plt.plot(steps, training_loss_gemma, marker='o', label='Gemma 3B - Training Loss')
plt.title('Training Loss Verlauf')
plt.xlabel('Training Steps')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Validation Loss
plt.figure(figsize=(10, 5))
plt.plot(steps, validation_loss_qwen, marker='o', label='Qwen2.5VL 3B - Validation Loss')
plt.plot(steps, validation_loss_gemma, marker='o', label='Gemma 3B - Validation Loss')
plt.title('Validation Loss Verlauf')
plt.xlabel('Training Steps')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
